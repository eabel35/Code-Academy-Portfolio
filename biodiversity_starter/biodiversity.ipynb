```python
spec.columns
spec.conservation_status.value_counts()
```

```python
#noticed a lot of values in spec.conservation_status are NaN, I want to display this for comparison
#I will change NaN to text to be recognized as a category
spec.conservation_status = spec.conservation_status.apply(lambda x: str(x))
spec.conservation_status.value_counts()
```

```python
#merge these tables on scientific name
mer = obs.merge(spec, how = 'inner', on = 'scientific_name')
#I like to add a agg_val column of arbitrary values to make my groupbys read better
mer['agg_val'] = 0
```

```python
#breakdown of amount of species of different statuses at each park, is there a relationship?
mer.groupby(['conservation_status','park_name']).agg({'agg_val':'count'})
```

```python
#seems to be an even breakdown of species of different categories across all parks, 
#probably intentional choice of data curator

#is there some relationship between number of observations and status?
#let me check what the average number of observations for a species given its status and location
mer.groupby(['conservation_status','park_name']).agg({'observations':'mean'})
```

```python
#so there generally are less observations of endangered species on average than other less severe categories
#this also gives me a picture of what the other categories might mean in terms of num of observations
#without doing research, it seems there are occurances of the plants in the gen order:
# no status>species of concern>in recovery>=threatened>endangered

```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#let's create some columns with categories stored as integers
#analyze with categories of park name, category, convervation status
mer
```

```python
multreg = pd.DataFrame()
multreg['park_name'] = mer.park_name.astype("category").cat.codes
multreg['category'] = mer.category.astype("category").cat.codes
multreg['conservation_status'] = mer.conservation_status.astype("category").cat.codes
multreg['observations'] = mer['observations']
multreg
```

```python
X=multreg.iloc[:,-3:-1].values
X
```

```python
y=multreg.iloc[:,-1:].values
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
model = LinearRegression().fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
