```python
import pandas as pd
import numpy as np
```

```python
obs = pd.read_csv('observations.csv')
spec = pd.read_csv('species_info.csv')
```

```python
obs.sort_values(by = 'observations', ascending = True).head(40)
```

```python
obs.groupby('scientific_name').agg({'observations':'sum'})
```

```python
spec
```

```python
mer = obs.merge(spec, how = 'inner', on = 'scientific_name')
```

```python
mer.groupby(['conservation_status','park_name']).agg({'observations':'count'})
```

```python
mer.conservation_status = mer.conservation_status.apply(lambda x: str(x))
```

```python
mer.loc[mer.conservation_status == 'nan', 'conservation_status'] = 'none'
```
